# League wide win-loss trends

```python
# Overall average number of wins per season
average_wins = df.groupby('year')['wins'].mean()
print("Average Wins Per Season: ")
print(average_wins)
```

```python
# plotting results
average_wins = df.groupby('year')['wins'].mean().reset_index()
plt.figure(figsize=(10, 6))
bars = plt.bar(average_wins['year'], average_wins['wins'], color = 'skyblue')
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval, round(yval, 2), ha = 'center', va = 'bottom')
plt.title("Average Wins By Season Across All Teams")
plt.xlabel('Year')
plt.ylabel('Average Wins')
plt.show()
```

The average number of wins across all teams remained very stagnant, but spiked from 7.97 in 2020, to 8.47 in 2021. I was curious about the percentage jump, so the code below will calculate the increase as a percentage.

```python
# calculating percentage
wins_2020 = 7.97
wins_2021 = 8.47

percent_jump = ((wins_2021 - wins_2020) / wins_2020) * 100

print(f"The percent increase from {wins_2020} to {wins_2021} is: {percent_jump:.2f}%")
```

```python
# calculating average losses per season
average_losses = df.groupby('year')['losses'].mean()
print("Average Losses Per Season: ")
print(average_losses)
```

```python
# plotting results
average_losses = df.groupby('year')['losses'].mean().reset_index()
plt.figure(figsize=(10, 6))
bars = plt.bar(average_losses['year'], average_losses['losses'], color = 'skyblue')
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval, round(yval, 2), ha = 'center', va = 'bottom')
plt.title("Average Losses By Season Across All Teams")
plt.xlabel('Year')
plt.ylabel('Average Losses')
plt.show()
```

Interesting. The average amount of losses per season exactly matches average amount of wins. This could either mean that the dataset is extremely correlated, or there are some data quality issues that need addressing. 

# Average Points Scored Per Season

```python
plt.figure(figsize = (18, 12))

# calculating average points scored by year(season)
points_scored_trend = df.groupby('year')['points_scored'].mean()

# plotting the trends
points_scored_trend.plot(marker = 'o')

# adding data values to each point in chart
for year, avg_points_scored in points_scored_trend.items():
    plt.text(year, avg_points_scored, f'{avg_points_scored:.2f}', ha = 'left', va = 'bottom')

# customizing the plot
plt.title("Average Points Scored Per Game Trends")
plt.xlabel('Year')
plt.ylabel('Average Points Scored')
plt.show()
```

The insight for this chart is that overall scoring had two major increases:

1.) From 2009 to 2013, scoring per season increased from 343.47 points to 374.53 points.

2.) From 2019 to 2020, scoring per season increased from 365 points to 396.62 points.

This can infer that offenses are simply becoming more effective and efficient, but could also point questions to how effective defense is, especially in this current era, which will be analyzed later in the analysis.

# Average Points Scored Per Game, By Team

```python
df['avg_points_per_game'] = df['points_scored'] / (df['wins'] + df['losses'])

# Group by Team and Calculate Average
average_points_by_team = df.groupby('team')['avg_points_per_game'].mean().reset_index()

# Sorting results in descending order
average_points_by_team = average_points_by_team.sort_values(by = 'avg_points_per_game', ascending = False)

# Display the result
print("Average Points Scored per Game in a Season for Each Team:")
print(average_points_by_team)

# Visualize the result
plt.figure(figsize=(18, 12))
plt.bar(average_points_by_team['team'], average_points_by_team['avg_points_per_game'], color='skyblue')
plt.title("Average Points Scored per Game by Each Team")
plt.xlabel('Team')
plt.ylabel('Average Points per Game')
plt.xticks(rotation=45, ha='right')
plt.show()
```

This visual shows the average points scored per game, across 20 seasons. The New England Patriots led the league, with an average of 26.85 points per game, followed by the New Orleans Saints(25.88 points per game) and the Green Bay Packers(25.68 points per game)

# Average Wins By Team Per Season

```python
# Calculate total wins per team
total_wins_per_team = df.groupby('team')['wins'].sum().reset_index()

# Calculate total years per team
total_years_per_team = df.groupby('team')['year'].nunique().reset_index()

# Merge the two DataFrames on the 'team' column
team_stats = pd.merge(total_wins_per_team, total_years_per_team, on='team', how='inner')

# Calculate average wins per year
team_stats['avg_wins_per_year'] = team_stats['wins'] / team_stats['year']

# Sort the result by avg_wins_per_year in descending order
team_stats = team_stats.sort_values(by='avg_wins_per_year', ascending=False)

# Display the result
print("Average Wins Per Year For Each Team:")
print(team_stats[['team', 'avg_wins_per_year']])
```

In this graphic, the most consistent teams are the Patriots and Steelers. It makes sense as these two franchises both have 6 Super Bowl titles.
